# F3 - Visualisation

Chargement des données et génération des graphiques demandés.

```

import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

df = pd.read_csv("data/train_clean.csv")

os.makedirs("figures", exist_ok=True)

sns.set_style("whitegrid")
```

## 18. Histogramme + KDE de Annual_Income selon Credit_Score

```

plt.figure(figsize=(10,6))

for score in df["Credit_Score"].unique():
    subset = df[df["Credit_Score"] == score]
    sns.histplot(
        subset["Annual_Income"],
        kde=True,
        stat="density",
        alpha=0.3,
        label=score
    )

plt.title("Distribution de Annual_Income selon Credit_Score")
plt.xlabel("Annual_Income")
plt.ylabel("Densité")
plt.legend()
plt.tight_layout()
plt.savefig("figures/hist_kde_annual_income.png")
plt.show()
```

## 19. Boxplots

```

plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Credit_Score", y="Delay_from_due_date")
plt.title("Delay_from_due_date selon Credit_Score")
plt.tight_layout()
plt.savefig("figures/boxplot_delay_due_date.png")
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Credit_Score", y="Outstanding_Debt")
plt.title("Outstanding_Debt selon Credit_Score")
plt.tight_layout()
plt.savefig("figures/boxplot_outstanding_debt.png")
plt.show()
```

## 20. Heatmap des corrélations

```

numeric_df = df.select_dtypes(include=np.number)

corr_matrix = numeric_df.corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Matrice de corrélation")
plt.tight_layout()
plt.savefig("figures/correlation_heatmap.png")
plt.show()

corr_pairs = corr_matrix.abs().unstack()
corr_pairs = corr_pairs[corr_pairs < 1]

top_corr = corr_pairs.sort_values(ascending=False).drop_duplicates().head(3)

print("Top 3 corrélations :")
for (v1, v2), corr in top_corr.items():
    print(f"{v1} <-> {v2} : {corr:.3f}")
```

## 21. Scatter Plot

```

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Annual_Income",
    y="Outstanding_Debt",
    hue="Credit_Score",
    alpha=0.7
)

plt.title("Annual_Income vs Outstanding_Debt")
plt.tight_layout()
plt.savefig("figures/scatter_income_debt.png")
plt.show()
```

## 22. Répartition de Credit_Score selon Payment_Behaviour

```

plt.figure(figsize=(14,7))

sns.countplot(
    data=df,
    x="Payment_Behaviour",
    hue="Credit_Score"
)

plt.xticks(rotation=45)
plt.title("Répartition de Credit_Score selon Payment_Behaviour")
plt.tight_layout()
plt.savefig("figures/barplot_payment_behaviour.png")
plt.show()
```

## 23. Test du Chi²

```

table = pd.crosstab(
    df["Payment_of_Min_Amount"],
    df["Credit_Score"]
)

chi2, p_value, dof, expected = chi2_contingency(table)

print(f"Chi² = {chi2:.4f}")
print(f"p-value = {p_value:.6f}")

alpha = 0.05

if p_value < alpha:
    print("Conclusion : rejet de H0 (variables dépendantes).")
else:
    print("Conclusion : on ne rejette pas H0 (variables indépendantes).")
```
