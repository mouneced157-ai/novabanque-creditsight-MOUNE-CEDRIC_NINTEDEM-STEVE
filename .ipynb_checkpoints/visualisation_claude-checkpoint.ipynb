# F3 : Visualisation — NovaBanque S.A.
**Fichier source :** `data/train_clean.csv`  
**Objectif :** Explorer et visualiser les données de credit scoring (tâches 18 à 23)

## Imports & Configuration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats

%matplotlib inline

# Dossier de sauvegarde des figures
FIGURES_DIR = "figures"
os.makedirs(FIGURES_DIR, exist_ok=True)

# Palette et ordre communs
credit_score_order = ["Poor", "Standard", "Good"]
palette = {"Poor": "#e74c3c", "Standard": "#f39c12", "Good": "#2ecc71"}

def save_fig(fig, name):
    path = os.path.join(FIGURES_DIR, name)
    fig.savefig(path, dpi=150, bbox_inches="tight")
    print(f"Sauvegardé → {path}")
```

## Chargement du dataset

```python
df = pd.read_csv("data/train_clean.csv", index_col=False)
print(f"Shape : {df.shape}")
df.head()
```

## Tâche 18 — Histogramme + KDE du `Annual_Income` par `Credit_Score`
3 courbes superposées (une par catégorie de Credit_Score), sauvegardées en PNG.

```python
fig, ax = plt.subplots(figsize=(10, 6))

for cs in credit_score_order:
    subset = df[df["Credit_Score"] == cs]["Annual_Income"].dropna()
    subset.plot.hist(ax=ax, bins=40, alpha=0.35, color=palette[cs],
                     density=True, label=f"{cs} (hist)")
    subset.plot.kde(ax=ax, color=palette[cs], linewidth=2, label=f"{cs} (KDE)")

ax.set_title("Distribution du Annual_Income par Credit_Score\n(Histogramme + KDE)", fontsize=14)
ax.set_xlabel("Annual Income")
ax.set_ylabel("Densité")
ax.legend()
plt.tight_layout()
save_fig(fig, "18_hist_kde_annual_income.png")
plt.show()
```

## Tâche 19 — Boxplots : `Delay_from_due_date` & `Outstanding_Debt` par `Credit_Score`

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, col in zip(axes, ["Delay_from_due_date", "Outstanding_Debt"]):
    order = [cs for cs in credit_score_order if cs in df["Credit_Score"].unique()]
    data_plot = [df[df["Credit_Score"] == cs][col].dropna() for cs in order]
    bp = ax.boxplot(data_plot, patch_artist=True, labels=order)
    for patch, cs in zip(bp["boxes"], order):
        patch.set_facecolor(palette[cs])
        patch.set_alpha(0.7)
    ax.set_title(f"{col}\npar Credit_Score", fontsize=12)
    ax.set_xlabel("Credit_Score")
    ax.set_ylabel(col)

fig.suptitle("Boxplots : Delay_from_due_date & Outstanding_Debt selon Credit_Score", fontsize=13)
plt.tight_layout()
save_fig(fig, "19_boxplot_delay_outstanding.png")
plt.show()
```

## Tâche 20 — Heatmap de corrélation (toutes variables numériques)
Commentaire des 3 corrélations les plus fortes.

```python
num_df = df.select_dtypes(include="number")
corr = num_df.corr()

fig, ax = plt.subplots(figsize=(14, 11))
sns.heatmap(corr, ax=ax, annot=True, fmt=".2f", cmap="coolwarm",
            center=0, linewidths=0.4, annot_kws={"size": 7})
ax.set_title("Heatmap de corrélation – variables numériques", fontsize=14)
plt.tight_layout()
save_fig(fig, "20_heatmap_correlation.png")
plt.show()
```

```python
# Top 3 corrélations (triangle supérieur uniquement)
corr_abs = corr.abs()
mask = pd.DataFrame(True, index=corr.index, columns=corr.columns)
for i in range(len(mask)):
    for j in range(i + 1):
        mask.iloc[i, j] = False

top3 = corr_abs.where(mask).stack().sort_values(ascending=False).head(3)

print("── Top 3 corrélations ──")
for (v1, v2), val in top3.items():
    direction = "positive" if corr.loc[v1, v2] > 0 else "négative"
    print(f"  {v1} <-> {v2} : r = {corr.loc[v1, v2]:.3f}  ({direction})")
```

## Tâche 21 — Scatter plot : `Annual_Income` vs `Outstanding_Debt` (coloré par `Credit_Score`)

```python
fig, ax = plt.subplots(figsize=(10, 7))

for cs in credit_score_order:
    sub = df[df["Credit_Score"] == cs]
    ax.scatter(sub["Annual_Income"], sub["Outstanding_Debt"],
               c=palette[cs], label=cs, alpha=0.4, s=15, edgecolors="none")

ax.set_title("Annual_Income vs Outstanding_Debt\ncoloré par Credit_Score", fontsize=13)
ax.set_xlabel("Annual Income")
ax.set_ylabel("Outstanding Debt")
ax.legend(title="Credit_Score")
plt.tight_layout()
save_fig(fig, "21_scatter_income_vs_debt.png")
plt.show()
```

## Tâche 22 — Barplot : répartition de `Credit_Score` selon `Payment_Behaviour`

```python
ct = pd.crosstab(df["Payment_Behaviour"], df["Credit_Score"], normalize="index") * 100

fig, ax = plt.subplots(figsize=(12, 6))
ct[credit_score_order].plot(kind="bar", ax=ax,
                             color=[palette[cs] for cs in credit_score_order],
                             edgecolor="white", width=0.7)
ax.set_title("Répartition du Credit_Score selon Payment_Behaviour (%)", fontsize=13)
ax.set_xlabel("Payment_Behaviour")
ax.set_ylabel("Proportion (%)")
ax.legend(title="Credit_Score")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
save_fig(fig, "22_barplot_credit_score_payment_behaviour.png")
plt.show()
```

## Tâche 23 — Test du χ² : `Payment_of_Min_Amount` vs `Credit_Score`
Seuil : **α = 0.05**

```python
contingency = pd.crosstab(df["Payment_of_Min_Amount"], df["Credit_Score"])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency)

print(f"χ²      = {chi2_stat:.4f}")
print(f"p-value = {p_value:.4e}")
print(f"ddl     = {dof}")

alpha = 0.05
if p_value < alpha:
    print(f"\n✅ p-value ({p_value:.2e}) < α ({alpha}) → On rejette H₀.")
    print("   Il existe une association statistiquement significative entre")
    print("   Payment_of_Min_Amount et Credit_Score.")
else:
    print(f"\n❌ p-value ({p_value:.2e}) ≥ α ({alpha}) → On ne rejette pas H₀.")
    print("   Pas d'association significative entre Payment_of_Min_Amount et Credit_Score.")
```
